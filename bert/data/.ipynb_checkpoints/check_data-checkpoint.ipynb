```python
import pickle

flipped_gap = pickle.load(open("./13436_flipped_gap_sents.pkl", 'rb'))
# ori_gap = pickle.load(open("./13436_ori_gap_sents.pkl", 'rb'))
```

```python
from transformers import BertTokenizer, BertModel
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

sents = []
for i,sent in enumerate(flipped_gap):
    if len(tokenizer.tokenize(sent)) > 128:
        sents.append(i)
```

```python
incorrects = [8, 9, 18, 41, 42, 45, 46, 70, 73, 74, 94, 97 ]
```

```python
import json

dic = dict({'incorrect' : [], 'correct' : []})
with open("quality_check/labeled_sent.json", 'w') as f:
    for i, sent in enumerate(cur):
        idx = i + 1
        if idx in incorrects:
            dic['incorrect'].append(sent)
        else:
            dic['correct'].append(sent)
    json.dump(dic, f, indent = '\t')
    
```
